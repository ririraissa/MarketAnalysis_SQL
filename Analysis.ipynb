# SQL

```python
import sqlite3
import locale
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')

# Connect to a SQLite database
conn = sqlite3.connect('my_database.db')

# Define a list of DataFrames and their corresponding table names
dataframes = [
    (customer, 'Customer'),
    (marketing, 'Marketing_Campaign'),
    (perusahaan, 'Review_Perusahaan'),
    (produk, 'Review_Produk'),
    (main, 'Transaksi_Main_Product'),
    (service, 'Transaksi_Service')
]

# Write each DataFrame to the SQLite database
for dataframe, table_name in dataframes:
    dataframe.to_sql(table_name, conn, index=False, if_exists='replace')
```

## Total Transaksi dan Revenue Product

```python
query = """
    select  Customer,
            count(ID_Transaksi) as Total_Transaksi,
            sum(Harga_Asli) as Revenue
    from Transaksi_Main_Product
    group by 1
    order by 3 desc
    limit 5;
"""
result = pd.read_sql_query(query, conn)
result['Revenue'] = result['Revenue'].apply(lambda x: '{:,.0f}'.format(x)) # Biar formatnya lebih enak dibaca

result
```

## Total Transaksi dan Revenue Tiap Kota

```python
query = """
    select  Customer.Domisili,
            count(Transaksi_Main_Product.ID_Transaksi) as Total_Transaksi,
            sum(Transaksi_Main_Product.Harga_Asli) as Revenue
    from Customer
    join Transaksi_Main_Product on Customer.Name = Transaksi_Main_Product.Customer
    group by 1
    order by 3 desc
    limit 5;
"""
result = pd.read_sql_query(query, conn)
result['Revenue'] = result['Revenue'].apply(lambda x: '{:,.0f}'.format(x)) # Biar formatnya lebih enak dibaca

result
```

## Total Revenue Produk + Service Tiap Bulan

```python
query = """
    with Gabung as (
        select Tanggal_Transaksi as Tgl, Harga_Asli as Price
        from Transaksi_Main_Product
        union
        select Tanggal_Service as Tgl, Harga_Service as Price
        from Transaksi_Service
    )
    select
        strftime('%m', Tgl) AS month,
        SUM(Price) AS Revenue
    from Gabung
    group by 1
    order by 1;
"""
result = pd.read_sql_query(query, conn)
result['Revenue'] = result['Revenue'].apply(lambda x: '{:,.0f}'.format(x)) # Biar formatnya lebih enak dibaca

result
```

## Produk Disukai Wanita umur 20 - 29 (top 3)

```python
query = """
WITH Cust AS (
    SELECT Name, Usia, Gender
    FROM Customer
    WHERE Usia >= 20 AND Usia <= 29 AND Gender = 'Wanita'
)
SELECT Tipe_Produk, COUNT(ID_Transaksi) AS Jumlah
FROM Transaksi_Main_Product
WHERE Customer IN (SELECT Name FROM Cust)
GROUP BY 1
ORDER BY 2 DESC
LIMIT 3;
"""
result = pd.read_sql_query(query, conn)

result
```

## Revenue dari Cust dengan Rating < 5 Perusahaan

```python
query = """
WITH Data_cust AS (
    SELECT Customer As Customer, Harga_Service As Price
    FROM Transaksi_Service
    union
    select Customer As Customer, Harga_Asli As Price
    from Transaksi_Main_Product
)
select Data_cust.Customer, sum(Data_cust.Price) as Revenue
from Data_cust
join Review_Perusahaan
on Data_cust.Customer = Review_Perusahaan.Name
where Review_Perusahaan.Rating < 5
group by 1;

"""
result = pd.read_sql_query(query, conn)
result['Revenue'] = result['Revenue'].apply(lambda x: '{:,.0f}'.format(x)) # Biar formatnya lebih enak dibaca

result
```

```python
query = """
WITH Data_cust AS (
    SELECT Customer As Customer, Harga_Service As Price
    FROM Transaksi_Service
    union
    select Customer As Customer, Harga_Asli As Price
    from Transaksi_Main_Product
)
select Data_cust.Customer, sum(Data_cust.Price) as Revenue
from Data_cust
join Review_Perusahaan
on Data_cust.Customer = Review_Perusahaan.Name
where Review_Perusahaan.Rating < 5
group by 1
order by 2 desc
limit 5;

"""
result = pd.read_sql_query(query, conn)
result['Revenue'] = result['Revenue'].apply(lambda x: '{:,.0f}'.format(x)) # Biar formatnya lebih enak dibaca

result
```

## Total Transaksi dan Revenue setiap Campaign

```python
query = """
WITH Gabung AS (
    SELECT Tanggal_Transaksi as Tgl, Harga_Asli as Price
    FROM Transaksi_Main_Product
    UNION
    SELECT Tanggal_Service as Tgl, Harga_Service as Price
    FROM Transaksi_Service
)
SELECT Marketing_Campaign.name,count(Gabung.Tgl) as Transactions, sum(Gabung.price) as Revenue
from Gabung
join Marketing_Campaign on Gabung.Tgl between Marketing_Campaign.start_date and Marketing_Campaign.end_date
group by Marketing_Campaign.name;
"""

result = pd.read_sql_query(query, conn)
result['Revenue'] = result['Revenue'].apply(lambda x: '{:,.0f}'.format(x)) # Biar formatnya lebih enak dibaca

result
```

```python
product_review = main.merge(produk, left_on='ID_Transaksi', right_on='ID_Review', how='left')
product_review
```

Coba lihat produk mana yang paling banyak dikasih rating

```python
product_review[product_review['Review'].isna()]['Tipe_Produk'].value_counts()
```

Coba lihat median rating tiap-tiap produk

```python
product_review.groupby('Tipe_Produk')['Review'].median()
```

Produk E kemungkinan ada masalah, karena banyak yang tidak mengisi rating serta median ratingnya rendah

```python
import matplotlib.pyplot as plt

# Memvisualisasikan data
product_e_data = product_review[product_review['Tipe_Produk'] == 'Produk E'].groupby('month')['Review'].mean()
product_e_data.plot()

# Memberi nama plot
plt.title('Rating Produk E untuk setiap Bulan')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Memvisualisasikan data
product_e_data = product_review[product_review['Tipe_Produk'] == 'Produk E'].groupby('month')['Harga_Asli'].sum()
product_e_data.plot()

# Memberi nama plot
plt.title('Revenue Produk E untuk tiap Bulan')

plt.show()
```

Setelah bulan 2 yang ratingnya tinggi, terjadi peningkatan revenue di bulan 3.

Namun revenue mulai turun ketika rating produk turun di bulan 3 dan seterusnya.

**Rekomendasi :**
* Perlu analisa lebih lanjut penyebab turunnya performa produk E dari bulan ke bulan
* Perlu branding ulang dan membangun kembali rating produk E jika ingin tetap mendapatkan revenue dari produk tersebut

## Performa Marketing

```python
month_budget = marketing.copy()
month_budget['month'] = month_budget['Name'].str.replace('Campaign ', '').astype(int)
month_budget = month_budget[['month', 'Budget']]
month_budget
```

```python
month_revenue = main[main['month'] <=7].groupby('month')['Harga_Asli'].sum().reset_index()
month_revenue
```

Sepertinya ada penurunan performa marketing, budget meningkat tidak diikuti dengan revenue yang tidak mempunyai efek yang signifikan.

Coba lihat persentase peningkatan revenue dari budget

```python
revenue_budget['percentage'].plot()
```
